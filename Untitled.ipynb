```python
import seaborn as sns
import matplotlib.pyplot as plt

# Função para dividir e visualizar a matriz de correlação por blocos de colunas
def plot_correlation_blocks(df, block_size=10):
    num_vars = df.shape[1]  # Número total de variáveis
    blocks = [df.columns[i:i + block_size] for i in range(0, num_vars, block_size)]  # Dividir em blocos

    for i, block in enumerate(blocks):
        fig, ax = plt.subplots(figsize=(10, 8))  # Tamanho da figura ajustado
        corr = df.corr(method="pearson")  # Matriz de correlação completa
        sns.heatmap(corr.loc[:, block], annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5, ax=ax)  # Apenas algumas colunas
        ax.set_title(f"Correlation Heatmap - Columns {i*block_size+1} to {(i+1)*block_size}")
        plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotacionar rótulos no eixo x
        plt.yticks(fontsize=10)
        plt.tight_layout()
        plt.show()

# Dividir a matriz de correlação em blocos de 10 colunas
plot_correlation_blocks(df[metric_features], block_size=10)


```
